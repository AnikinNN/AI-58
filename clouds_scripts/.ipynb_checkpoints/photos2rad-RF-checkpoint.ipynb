```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model, model_selection, neural_network, preprocessing
from datetime import datetime
```

```python
def mape(y_true, y_pred):
    return np.mean(np.abs((y_pred-y_true)/(y_true+(1e-8))))
```

```python
from service_defs import *
```

```python
ds_fnames = find_files('./../clouds_databases', 'dataset-*.csv')
```

```python
ds_fnames
```

```python
dfs = [pd.read_csv(fn, sep=',', decimal='.', parse_dates=['photo_time', 'data time']) for fn in ds_fnames]
```

```python
df = pd.concat(dfs)
df = df.reset_index(drop=True)
```

```python
df.shape
```

```python
df
```

```python
df['date-hour'] = df.photo_time.apply(lambda dt: str(dt.date())+'-'+('%02d'%dt.hour))
```

```python
feature_columns = [c for c in df.columns if c.startswith('f')]
```

```python
df = df[df['CM3up[W/m2]'] >= 5.0]
```

```python
df.shape
```

```python
dates = np.unique(df['date-hour'])
dates.shape
```

```python
dates_train, dates_test = train_test_split(dates, test_size=0.25)
```

```python
ytr = df[df['date-hour'].isin(dates_train)]['CM3up[W/m2]'].values
ytest = df[~df['date-hour'].isin(dates_train)]['CM3up[W/m2]'].values
```

```python
Xtr = df[df['date-hour'].isin(dates_train)][feature_columns].values
Xtest = df[~df['date-hour'].isin(dates_train)][feature_columns].values
```

```python
scaler = preprocessing.StandardScaler()

Xtr = scaler.fit_transform(Xtr)
Xtest = scaler.fit_transform(Xtest)
```

```python
Xtr.shape, Xtest.shape
```

```python
model = neural_network.MLPRegressor(hidden_layer_sizes=[256, 256, 256],
                                    verbose=True,
                                    max_iter=400,
                                    learning_rate_init=1e-4,
                                 )

# model = linear_model.LinearRegression()
```

```python
model = model_selection.GridSearchCV(neural_network.MLPRegressor(max_iter = 1000, learning_rate_init=1e-4, verbose=True,),
                                     param_grid=
                                         {
                                             "hidden_layer_sizes": [
                                                 [256] * 3,
                                                 [256] * 4,
                                                 [256] * 5,
                                                 [512] * 3,
                                                 [512] * 4,
                                                 [512] * 5,
                                             ],
                                             "activation": ["relu", "logistic"]
                                         },
                                     n_jobs=6,
                                     verbose=10,
                                    )
```

```python
timer = datetime.datetime.now()
_ = model.fit(Xtr, ytr)
timer = datetime.datetime.now() - timer
print(timer)
```

```python
model.
```

```python
model.best_estimator_
```

```python
ytr_pred = model.best_estimator_.predict(Xtr)
```

```python
np.sqrt(mean_squared_error(ytr, ytr_pred)), mean_absolute_error(ytr, ytr_pred), mape(ytr, ytr_pred)
```

```python
vmin = 5.0
vmax = max(np.percentile(ytr, 99), np.percentile(ytest, 99))
bins = np.linspace(vmin, vmax, 100, endpoint=True)
```

```python
f = plt.figure(figsize=(6,4), dpi=100)
_ = plt.hist(ytr, bins=bins, density=True, alpha=0.7, label='train label')
_ = plt.hist(ytest, bins=bins, density=True, alpha=0.7, label='test label')
_ = plt.legend()
```

```python
f = plt.figure(figsize=(6,4), dpi=100)
_ = plt.hist(np.sign(ytr_pred-ytr)*np.log(np.abs((ytr_pred-ytr+1e-8))), bins=100)
```
